```python
# Importing necessary libraries
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Reading the dataset into a DataFrame
df = pd.read_csv('USvideos.csv')
```

```python
# Displaying the first five rows of the dataset to understand its structure
df.head()
```

```python
# Displaying the shape of the dataset to understand the number of rows and columns
df.shape
```

```python
df = df.drop_duplicates()  # Dropping duplicate rows from the 'df' dataframe.
df.shape  # Calculating the current shape (number of rows and columns) of the 'df' dataframe.
```

```python
df.describe()  # Generating descriptive statistics of the 'df' dataframe.
```

```python
df.info()  # Providing concise summary information about the 'df' dataframe.
```

```python
columns_to_remove = ['thumbnail_link', 'description']
df = df.drop(columns=columns_to_remove)  # Dropping specified columns ('thumbnail_link', 'description') from the 'df' dataframe.
df.info()  # Providing concise summary information about the modified 'df' dataframe.
```

```python
from datetime import datetime  # Importing datetime module from the datetime package
import datetime

# Converting 'trending_date' column to datetime format using lambda function and datetime.strptime
df["trending_date"] = df["trending_date"].apply(lambda x: datetime.datetime.strptime(x, '%y.%d.%m'))
df.head(3)  # Displaying the first 3 rows of the 'df' dataframe after converting 'trending_date' to datetime format.
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])  # Converting 'publish_time' column to datetime format using pandas.
df.head(2)  # Displaying the first 2 rows of the 'df' dataframe after converting 'publish_time' to datetime.
```

```python
df['publish_month'] = df['publish_time'].dt.month  # Extracting month from 'publish_time' column and storing in 'publish_month'
df['publish_day'] = df['publish_time'].dt.day  # Extracting day from 'publish_time' column and storing in 'publish_day'
df['publish_hour'] = df['publish_time'].dt.hour  # Extracting hour from 'publish_time' column and storing in 'publish_hour'
df.head(2)  # Displaying the first 2 rows of the 'df' dataframe after adding 'publish_month', 'publish_day', and 'publish_hour'
```

```python
print(sorted(df["category_id"].unique()))  # Printing sorted unique values of 'category_id' column in the 'df' dataframe
```

```python
df['category_name'] = np.nan  # Creating a new column 'category_name' filled with NaN values

# Assigning category names based on category IDs
df.loc[df["category_id"] == 1, "category_name"] = 'Film and Animation'
df.loc[df["category_id"] == 2, "category_name"] = 'Autos and Vehicles'
df.loc[df["category_id"] == 10, "category_name"] = 'Music'
df.loc[df["category_id"] == 15, "category_name"] = 'Pets and Animals'
df.loc[df["category_id"] == 17, "category_name"] = 'Sports'
df.loc[df["category_id"] == 19, "category_name"] = 'Travel and Events'
df.loc[df["category_id"] == 20, "category_name"] = 'Gaming'
df.loc[df["category_id"] == 22, "category_name"] = 'People and Blogs'
df.loc[df["category_id"] == 23, "category_name"] = 'Comedy'
df.loc[df["category_id"] == 24, "category_name"] = 'Entertainment'
df.loc[df["category_id"] == 25, "category_name"] = 'News and Politics'
df.loc[df["category_id"] == 26, "category_name"] = 'How to and Style'
df.loc[df["category_id"] == 27, "category_name"] = 'Education'
df.loc[df["category_id"] == 28, "category_name"] = 'Science and Technology'
df.loc[df["category_id"] == 29, "category_name"] = 'Non Profits and Activism'
df.loc[df["category_id"] == 30, "category_name"] = 'Movies'
df.loc[df["category_id"] == 43, "category_name"] = 'Shows'

df.head()  # Displaying the first five rows of the 'df' dataframe after adding 'category_name'
```

```python
df['year'] = df['publish_time'].dt.year  # Extracting year from 'publish_time' column and storing in 'year'

yearly_counts = df.groupby('year')['video_id'].count()  # Grouping by 'year' and counting total 'video_id' per year
yearly_counts.plot(kind='bar', xlabel='Year', ylabel='Total Publish Count', title='Total Publish Video Per Year')
plt.show()  # Displaying the bar plot showing total publish video count per year
```

```python
df['year'] = df['publish_time'].dt.year  # Extracting year from 'publish_time' column and storing in 'year'

yearly_views = df.groupby('year')['views'].sum()  # Grouping by 'year' and summing 'views' per year
yearly_views.plot(kind='bar', xlabel='Year', ylabel='Total views', title='Total Views per Year')

plt.xticks(rotation=8)  # Rotating x-axis labels for better readability
plt.tight_layout()  # Ensuring tight layout for better spacing
plt.show()  # Displaying the bar plot showing total views per year
```

```python
category_views = df.groupby('category_name')['views'].sum().reset_index()  # Grouping by 'category_name' and summing 'views', then resetting index
top_categories = category_views.sort_values(by='views', ascending=False).head(5)  # Sorting by 'views' in descending order and selecting top 5

plt.bar(top_categories['category_name'], top_categories['views'])  # Creating a bar plot with 'category_name' on x-axis and 'views' on y-axis
plt.xlabel('Category Name', fontsize=12)  # Setting x-axis label with fontsize 12
plt.ylabel('Total Views', fontsize=12)  # Setting y-axis label with fontsize 12
plt.title('Top 5 Categories', fontsize=15)  # Setting plot title with fontsize 15
plt.tight_layout()  # Ensuring tight layout for better spacing
plt.show()  # Displaying the bar plot showing top 5 categories by total views
```

```python
plt.figure(figsize=(12, 6))  # Setting figure size to 12 inches width and 6 inches height

# Creating a count plot with 'category_name' on x-axis using seaborn
sns.countplot(x='category_name', data=df, order=df['category_name'].value_counts().index)

plt.xticks(rotation=90)  # Rotating x-axis labels by 90 degrees for better readability
plt.title("Video Count by Category")  # Setting plot title
plt.show()  # Displaying the count plot showing video count by category
```

```python
videos_per_hour = df['publish_hour'].value_counts().sort_index()  # Counting videos per hour and sorting by hour

plt.figure(figsize=(12, 6))  # Setting figure size to 12 inches width and 6 inches height

# Creating a bar plot with 'publish_hour' on x-axis and count on y-axis using seaborn
sns.barplot(x=videos_per_hour.index, y=videos_per_hour.values, palette="rocket")

plt.title('Number of Videos Published per Hour')  # Setting plot title
plt.xlabel('Hour of Day')  # Setting x-axis label
plt.ylabel('Number of Videos')  # Setting y-axis label
plt.xticks(rotation=45)  # Rotating x-axis labels by 45 degrees for better readability
plt.show()  # Displaying the bar plot showing number of videos published per hour
```

```python
df['publish_time'] = pd.to_datetime(df['publish_time'])  # Converting 'publish_time' column to datetime format
df['publish_date'] = df['publish_time'].dt.date  # Extracting date from 'publish_time' and storing in 'publish_date'

video_count_by_date = df.groupby('publish_date').size()  # Counting number of videos published per date

plt.figure(figsize=(12, 6))  # Setting figure size to 12 inches width and 6 inches height

# Creating a line plot with 'publish_date' on x-axis and video count on y-axis using seaborn
sns.lineplot(data=video_count_by_date)

plt.title('Videos Published Over Time')  # Setting plot title
plt.xlabel('Publish Date')  # Setting x-axis label
plt.ylabel('Number of Videos')  # Setting y-axis label
plt.xticks(rotation=45)  # Rotating x-axis labels by 45 degrees for better readability
plt.show()  # Displaying the line plot showing videos published over time
```

```python
plt.figure(figsize=(10, 6))  # Setting figure size to 10 inches width and 6 inches height

# Creating a scatter plot with 'views' on x-axis and 'likes' on y-axis using seaborn
sns.scatterplot(data=df, x='views', y='likes')

plt.title('Views vs Likes')  # Setting plot title
plt.xlabel('Views')  # Setting x-axis label
plt.ylabel('Likes')  # Setting y-axis label
plt.show()  # Displaying the scatter plot showing relationship between views and likes
```

```python
plt.figure(figsize=(14, 8))  # Setting figure size to 14 inches width and 8 inches height
plt.subplots_adjust(wspace=0.2, hspace=0.4, top=0.9)  # Adjusting subplot spacing and top margin

# Subplot 1: Comments Disabled
plt.subplot(2, 2, 1)
g = sns.countplot(x='comments_disabled', data=df)
g.set_title("Comments Disabled", fontsize=16)

# Subplot 2: Ratings Disabled
plt.subplot(2, 2, 2)
g1 = sns.countplot(x='ratings_disabled', data=df)
g1.set_title("Rating Disabled", fontsize=16)

# Subplot 3: Video Error or Removed
plt.subplot(2, 2, 3)
g2 = sns.countplot(x='video_error_or_removed', data=df)
g2.set_title("Video Error or Removed", fontsize=16)

plt.show()  # Displaying the subplots
```
