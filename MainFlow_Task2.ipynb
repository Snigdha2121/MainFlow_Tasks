```python
# Importing the necessary library pandas as pd
import pandas as pd
```

```python
# RESPONSIBILITY 1: Loading the CSV file into a Pandas DataFrame
data = pd.read_csv("01.Data Cleaning and Preprocessing.csv")
```

```python
# Removing duplicate rows from the DataFrame
data = data.drop_duplicates()

data # It displays the contents of the DataFrame named 'data' after removing the duplicates
```

```python
# RESPONSIBILITY 2: Filtering data based on conditions

filtered_data = data[data['ChipRate'] > 16]  # Filter rows where 'ChipRate' is greater than 16
filtered_data # Displays the content after filtering data based on conditions. Here all the rows where 'Chiprate' was greater than 16 got printed. 
```

```python
# RESPONSIBILITY 3: Handling missing values

# This checks for missing values in the DataFrame. If there is a missing value it returns True else False. 
data.isnull()
```

```python
# This count the number of missing values in each column of the DataFrame
data.isnull().sum()
```

```python
# It check for non-null values in the DataFrame. If there exits a missing value it returns False, else it returns True. 
data.notnull()
```

```python
# It calculates the total number of missing values in the DataFrame
data.isnull().sum().sum()
```

```python
# Filling the missing values in the DataFrame with 0
data2 = data.fillna(value=0)

# Displaying the DataFrame after filling missing values
data2


# NOTE: There are 2 other methods to fill missing values:
# 1. Forward filling (ffill()): Fills missing values with the last observed value in the column.
# 2. Backward filling (bfill()): Fills missing values with the next observed value in the column.
# Example: data3 = data.ffill()  # Forward fill missing values
```

```python
data2.isnull().sum().sum() # Calculates the total number of missing values in the DataFrame data2. 
```

```python
# RESPONSIBILTY 4: Calculating summary statistics.

summary_stats = data2.describe()  # Generate descriptive statistics
summary_stats  # Dispalys the summary statistics
```
