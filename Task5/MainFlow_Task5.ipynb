```python
# Importing all the necessary libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Read the CSV file 'heart.csv' into a DataFrame
# The DataFrame, df, will contain the data from the CSV file
df = pd.read_csv('heart.csv')
```

```python
# Display the first five rows of the DataFrame
# This is useful for getting a quick overview of the data structure and content
df.head()
```

```python
# Display the last five rows of the DataFrame
# This helps to check the end portion of the dataset and ensure data integrity
df.tail()
```

```python
# Retrieve the column names of the DataFrame
# This provides an array of all column names present in the dataset
df.columns.values
```

```python
# Check for missing values in each column of the DataFrame
# df.isna() returns a DataFrame of boolean values indicating where NaN values are present
# df.isna().sum() calculates the sum of missing values (NaN) in each column
df.isna().sum()
```

```python
# Display a concise summary of the DataFrame
df.info()
```

```python
# Plotting histograms for each feature in the dataset
# This provides an overview of the distribution of each variable.

df.hist(bins=50, grid=False, figsize=(20, 15))
plt.suptitle('Distribution of Features', fontsize=20)  # Adding a title
plt.show()
```

```python
# Generate descriptive statistics of the numerical columns in the DataFrame
# df.describe() computes statistical summary like count, mean, standard deviation, min, quartiles, and max
df.describe()
```

```python
# List of questions related to the analysis of heart disease data

questions = [
    "1. How many people have heart disease and how many people doesn't have heart disease?",
    "2. People of which sex has most heart disease?",
    "3. People of which sex has which type of chest pain most?",
    "4. People with which chest pain are most prone to have heart disease?"]

questions
```

```python
# Question 1
# 1. How many people have heart disease and how many people doesn't have heart disease?

# Count the occurrences of each unique value in the 'target' column
# df.target.value_counts() returns a Series containing counts of unique values
# In the context of heart disease data, 'target' typically represents the presence (1) or absence (0) of heart disease
df.target.value_counts()
```

```python
# Plotting a bar chart to show the count of heart disease cases and non-cases.
df.target.value_counts().plot(kind='bar', color=["skyblue", "pink"])

# Setting the title of the plot
plt.title("Heart Disease values")

# Labeling the x-axis
plt.xlabel("1 Heart Disease, 0 No heart Disease")

# Labeling the y-axis
plt.ylabel("Amount")
```

```python
# Plotting a pie chart to show the proportion of heart disease cases vs. non-cases.
df.target.value_counts().plot(kind='pie', figsize=(8, 6))

# Adding a legend to the plot to label the categories
plt.legend(["Disease", "No disease"])

# Displaying the plot
plt.show()
```

```python
# Count the occurrences of each unique value in the 'sex' column
# df.sex.value_counts() returns a Series containing counts of unique values in the 'sex' column
# In the context of the dataset, typically 0 represents 'Female' and 1 represents 'Male'

df.sex.value_counts()
```

```python
# Plotting a pie chart of the counts of each sex

df.sex.value_counts().plot(kind='pie', figsize=(8, 6))  # Plotting the pie chart
plt.title('Male Female ratio')  # Setting the title of the plot
plt.legend(['Male', 'Female'])  # Adding a legend with labels for each category
plt.show()  # Display the plot
```

```python
# Question 2
# 2. People of which sex has most heart disease?

# Creating a cross-tabulation (contingency table) between 'target' and 'sex' columns
# pd.crosstab(df.target, df.sex) computes a frequency table that shows how many males and females have or do not have heart disease
cross_tab = pd.crosstab(df.target, df.sex)
```

```python
# Creating a count plot of 'target' with respect to 'sex'

# sns.countplot(x='target', data=df, hue='sex') creates a count plot where 'x' is 'target', 'data' is df, and 'hue' is 'sex'
sns.countplot(x='target', data=df, hue='sex')

plt.title("Heart Disease Frequency for Sex")  # Setting the title of the plot
plt.xlabel("0 = No heart Disease, 1 = Heart Disease")  # Setting the label for the x-axis
plt.show()  # Display the plot
```

```python
# Statement/Inference about the dataset observations

# Number of males is more than double that of females in our dataset.
# More than 45% of males have heart disease, while approximately 75% of females have heart disease.
```

```python
# Question 3
# 3. People of which sex has which type of chest pain most?

# Counting the occurrences of different types of chest pain (cp)
chest_pain_counts = df.cp.value_counts()
chest_pain_counts
```

```python
# Plotting a bar chart to visualize the counts of different chest pain types
# The plot displays each type of chest pain with a designated color

plt.figure(figsize=(8, 6))  # Creating a figure with specified size
df.cp.value_counts().plot(kind='bar', color=['blue', 'orange', 'green', 'red'])  # Plotting the bar chart
plt.title('Chest pain type count')  # Setting the title of the plot
plt.xlabel('Chest pain type')  # Setting the label for the x-axis
plt.ylabel('Count')  # Setting the label for the y-axis

plt.show()  # Display the plot
```

```python
# Creating a cross-tabulation (contingency table) between 'sex' and 'cp' columns

# pd.crosstab(df.sex, df.cp) computes a frequency table that shows the distribution of chest pain types across sexes
cross_tab = pd.crosstab(df.sex, df.cp)
cross_tab
```

```python
# Plotting a bar chart to visualize the distribution of chest pain types by sex
# The plot shows each type of chest pain categorized by sex, using specified colors for each type

plt.figure(figsize=(10, 6))  # Creating a figure with specified size
pd.crosstab(df.sex, df.cp).plot(kind='bar', color=['blue', 'orange', 'green', 'red'])  # Plotting the bar chart
plt.title('Type of chest pain for sex')  # Setting the title of the plot
plt.xlabel('Sex (0 = Female, 1 = Male)')  # Setting the label for the x-axis
plt.ylabel('Count')  # Setting the label for the y-axis
plt.legend(title='Chest Pain Type')  # Adding a legend with labels for each chest pain type
plt.xticks(rotation=0)  # Rotating x-axis labels for better readability

plt.show()  # Display the plot
```

```python
# Statement/Inference about the distribution of chest pain types by sex

# Most males have 'type 0' chest pain, and the least have 'type 4' pain.
# In the case of females, 'type 0' and 'type 2' percentages are almost the same.
```

```python
# Question 4
# 4. People with which chest pain are most pron to have heart disease?

# Creating a cross-tabulation (contingency table) between 'cp' (chest pain) and 'target' (heart disease presence)
# pd.crosstab(df.cp, df.target) computes a frequency table showing how many people with each type of chest pain have or do not have heart disease
chest_pain_target_cross = pd.crosstab(df.cp, df.target)

chest_pain_target_cross
```

```python
# Plotting a count plot to visualize the distribution of heart disease by chest pain type

# The plot shows the count of individuals with and without heart disease for each chest pain type

sns.countplot(x='cp', data=df, hue='target')  # Plotting the count plot
plt.title('Heart Disease by Chest Pain Type')  # Setting the title of the plot
plt.xlabel('Chest Pain Type')  # Setting the label for the x-axis
plt.ylabel('Count')  # Setting the label for the y-axis
plt.legend(title='Target', labels=['No Heart Disease', 'Heart Disease'])  # Adding a legend with labels for each target category
plt.show()  # Display the plot
```

```python
# Statement about chest pain types and heart disease
# Most people with type '0' chest pain have a lower chance of heart disease, whereas other types show the opposite trend.

# Now let's take a look at our age column.


# Creating a distribution plot with a normal distribution curve (Kernel Density Estimate, kde=True)
# The plot visualizes the distribution of age in the dataset, with 30 bins and a KDE plot for the density estimation

sns.histplot(x='age', data=df, bins=30, kde=True)  # Plotting the distribution plot with KDE
plt.title('Distribution of Age')  # Setting the title of the plot
plt.xlabel('Age')  # Setting the label for the x-axis
plt.ylabel('Frequency')  # Setting the label for the y-axis
plt.show()  # Display the plot
```

```python
# Statement/Inference about dataset characteristics
# 58-59 year old people are the most represented in the dataset.
# Let's plot another distribution plot for Maximum heart rate (thalach).


# Creating a distribution plot with a normal distribution curve (Kernel Density Estimate, kde=True)
# The plot visualizes the distribution of maximum heart rate (thalach) in the dataset

plt.figure(figsize=(10, 6))  # Creating a figure with specified size
sns.histplot(x='thalach', data=df, bins=30, kde=True, color='brown')  # Plotting the distribution plot with KDE
plt.title('Distribution of Maximum Heart Rate')  # Setting the title of the plot
plt.xlabel('Maximum Heart Rate (thalach)')  # Setting the label for the x-axis
plt.ylabel('Frequency')  # Setting the label for the y-axis
plt.show()  # Display the plot
```
