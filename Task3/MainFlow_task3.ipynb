```python
# Importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading the data from the CSV file into a pandas DataFrame
data = pd.read_csv('householdtask3.csv')
```

```python
# Displaying the first 10 rows of the DataFrame to inspect the initial data
data.head(10)
```

```python
# RESPONSIBILITY 1: SCATTER PLOTS

# Creating a scatter plot for 'year' against 'own' column from the DataFrame
plt.scatter(data['year'], data['own'])

# Setting the title of the scatter plot
plt.title("Scatter Plot for Year Against Own")

# Setting the label for the x-axis
plt.xlabel('Year')

# Setting the label for the y-axis
plt.ylabel('Own')

# Displaying the plot
plt.show()
```

```python
# Scatter plot for 'Age' against 'Income'
plt.scatter(data['age'], data['income'])

# Setting the title of the scatter plot
plt.title("Scatter Plot for Age Against Income")

# Setting the label for the x-axis
plt.xlabel('Age')

# Setting the label for the y-axis
plt.ylabel('Income')

# Displaying the plot
plt.show()
```

```python
# RESPONSIBILITY 2: LINE CHARTS

# Plotting 'year' column from the DataFrame on the x-axis
plt.plot(data['year'], label='Year')

# Plotting 'own' column from the DataFrame on the y-axis
plt.plot(data['own'], label='Own')

# Setting the title of the line chart
plt.title("Line Chart for Year and Own")

# Setting the label for the x-axis
plt.xlabel('Year')

# Setting the label for the y-axis
plt.ylabel('Own')

# Adding a legend to distinguish between the 'year' and 'own' lines
plt.legend()

# Displaying the plot
plt.show()
```

```python
# Plotting 'age' column from the DataFrame on the x-axis
plt.plot(data['age'], label='Age')

# Plotting 'income' column from the DataFrame on the y-axis
plt.plot(data['income'], label='Income')

# Setting the title of the line chart
plt.title("Line Chart for Age and Income")

# Setting the label for the x-axis
plt.xlabel('Age')

# Setting the label for the y-axis
plt.ylabel('Income')

# Adding a legend to distinguish between the 'age' and 'income' lines
plt.legend()

# Displaying the plot
plt.show()
```

```python
#RESPONSIBILTY 3: BAR CHART/GRAPH

# Plotting a bar chart of the 'own' column against the 'year' column from the DataFrame
plt.bar(data['year'], data['own'])

# Setting the title of the bar chart
plt.title("Bar Chart for Year and Own")

# Setting the label for the x-axis
plt.xlabel('Year')

# Setting the label for the y-axis
plt.ylabel('Own')

# Displaying the plot
plt.show()
```

```python
# Plotting a bar chart of the 'Age' column against the 'Income' column from the DataFrame
plt.bar(data['age'], data['income'])

# Setting the title of the bar chart
plt.title("Bar Chart for Age and Income")

# Setting the label for the x-axis
plt.xlabel('Age')

# Setting the label for the y-axis
plt.ylabel('Income')

# Displaying the plot
plt.show()
```

```python
# RESPOSIBILTY 4: CREATING HISTOGRAMS

# Plotting a histogram of the 'income' column from the DataFrame
plt.hist(data['income'])

# Setting the title of the histogram
plt.title("Histogram of Income")

# Setting the label for the x-axis
plt.xlabel('Income')

# Setting the label for the y-axis
plt.ylabel('Frequency')

# Displaying the plot
plt.show()
```
